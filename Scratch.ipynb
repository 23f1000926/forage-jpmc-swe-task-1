```python
text = input()
text = text.lower()
vowels = ""
if 'a' in text:
  vowels += 'a'
if 'e' in text:
  vowels += 'e'
if 'i' in text:
  vowels += 'i'
if 'o' in text:
  vowels += 'o'
if 'u' in text:
  vowels += 'u'
if vowels == "":
  print("none")
else:
  print(sorted(vowels), sep="")
```
